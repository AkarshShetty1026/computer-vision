```python
def image_manipulation (choice,label):
    import matplotlib.pyplot as plt
    import cv2 as cv
    import numpy as np

    image= cv.imread("ronaldo and messi.jpeg")


    if choice == 1:
        pt1_x=int(input("enter the x value of first point of the line: "))
        pt1_y=int(input("enter the y value of first point of the line: "))
        pt2_x=int(input("enter the x value of second point of the line: "))
        pt2_y=int(input("enter the y value of second point of the line: "))
        red=int(input("enter the red value of the color: "))
        green=int(input("enter the green value of the color: "))
        blue=int(input("enter the blue value of the color: "))

        result=cv.line(image,(pt1_x,pt1_y),(pt2_x,pt2_y),(blue,green,red))
    elif choice ==2:
        pt1_x=int(input("enter the x value of first point of the rectangle: "))
        pt1_y=int(input("enter the y value of first point of the rectangle: "))
        pt2_x=int(input("enter the x value of second point of the rectangle: "))
        pt2_y=int(input("enter the y value of second point of the rectangle: "))
        red=int(input("enter the red value of the color: "))
        green=int(input("enter the green value of the color: "))
        blue=int(input("enter the blue value of the color: "))
        filled=input("would you like the ractangle to be filled?: ")
        filled=filled.lower()
        if filled == "yes":
            thickness=-1
        else:
            thickness=3

        result=cv.rectangle(image,(pt1_x,pt1_y),(pt2_x,pt2_y),(blue,green,red), thickness=thickness)

    elif choice==3:
        centre_x=int(input("enter the x value of the centre of the circle: "))
        centre_y=int(input("enter the y value of the centre of the circle: "))
        radius=int(input("enter the radius of the circle: "))
        red=int(input("enter the red value of the color: "))
        green=int(input("enter the green value of the color: "))
        blue=int(input("enter the blue value of the color: "))
        filled=input("would you like the circle to be filled?: ")
        filled=filled.lower()
        if filled == "yes":
            thickness=-1
        else:
            thickness=3

        result=cv.circle(image,(centre_x,centre_y),radius,(blue,green,red),thickness=thickness)

    elif choice== 4:
        height=int(input("enter the height you want to resize the image to: "))
        width=int(input("enter the width you want to resize the image to: "))

        result=cv.resize(image,(width,height))


    elif choice==5:
        centre_x=int(input("enter the x value of the centre of the rotation: "))
        centre_y=int(input("enter the y value of the centre of the rotation: "))
        angle=int(input("enter at what angle you want to rotate to: "))
        scale=int(input("enter the scale factor: "))

        height=int(input("enter the height you want to resize the image to: "))
        width=int(input("enter the width you want to resize the image to: "))

        matrix=cv.getRotationMatrix2D((centre_x,centre_y),angle,scale)

        result=cv.warpAffine(image,matrix,(width,height))

    elif choice==6:
        print("1. flip vertically \n2. flip horizantally \n3. flip vertically and horizantally \n")
        flip_choice=int(input("enter which option on how you want to flip the image: "))

        if flip_choice == 1:
            flip_code=0
        elif flip_choice == 2:
            flip_code=1
        elif flip_choice == 3:
            flip_code=-1

        result=cv.flip(image,flip_code)

    elif choice ==7:
        rows= int(input("enter by how many rows you want to move the image: "))
        columns= int(input("enter by how many columns you want to move the image: "))

        height=int(input("enter the height you want to resize the image to: "))
        width=int(input("enter the width you want to resize the image to: "))

        array=np.float32([[1,0,rows],[0,1,columns]])

        result=cv.warpAffine(image,array,(width,height))

    elif choice ==8:
        result= cv.cvtColor(image,cv.COLOR_BGR2GRAY)

    elif choice ==9:
        thresh_value= int(input("enter the threshold value: "))
        maxval= int(input("enter the value of the maxvalue: "))
        gray= cv.cvtColor(image,cv.COLOR_BGR2GRAY)
        val,result= cv.threshold(gray,thresh_value,maxval,cv.THRESH_BINARY)  
    elif choice == 10:
        diameter=int(input("enter the diamater of the pixel neighbourhood: "))
        sigma_color=int(input("enter the sigma color value: "))
        sigma_space=int(input("enter the sigma space value: "))
        result=cv.bilateralFilter(image,diameter,sigma_color,sigma_space)
 
    #PLOTTING TO DISPLAY THE OUTPUT USING MATPLOTLIB

    result=cv.cvtColor(result,cv.COLOR_BGR2RGB)# this line is to convert the image pixel values form the BGR order to RGB as matplotlib reads images in the order RGB whereas cv2 reads it in the order BGR    
    plt.title(label)# this line is to label each image we plot
    plt.imshow(result)# this "imshow" function is from matplotlib and works in the same way as cv2's but it is not from cv2
    plt.show()# this line is used to display the plotted image
```

```python
options=["line","rectangle","circle","resize","rotate","flip","translation","grayscale","black and white","edge detection"]

for x in range(0,10):
    print(str(x+1)+".",options[x])
    
choice=int(input("enter which option you want: "))
label=str(options[choice-1])
image_manipulation(choice,label)
    
```
